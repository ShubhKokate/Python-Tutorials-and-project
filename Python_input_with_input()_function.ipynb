# Python input with input() Function

- In this tutorial, we will learn about the built-in _input()_ function.

- Up until now, our programs were static. The value of variables was defined or hard coded into the source code.

- To allow flexibility, we might want to take the input from the user.

- In Python, we have the _input()_ function to allow this.

- The syntax for input() is:

        > input([prompt])

- where _prompt_ is the string we wish to display on the screen. It is optional.

```python
num = input('Enter a number: ')
```

```python
num
```

- Here, we can see that the entered value 12 is a string, not a number.

- To convert this into a number we can use _int()_ or _float()_ functions.

- But eval takes it further. 

- It can evaluate even expressions, provided the input is a string.

```python
int('2+3')
```

```python
eval('2+3')

```
